# Python Boolean Data Type

## Summary

In this tutorial, you'll learn about the Python boolean data type, including the concepts of falsy and truthy values.

## Introduction to Python Boolean Data Type

In programming, it's common to check if a condition is true or not and perform actions based on the result. Python provides the boolean data type, denoted as `bool`, to represent true and false values.

The boolean data type has two values: `True` and `False`. It's essential to note that these values start with capital letters (`T` and `F`).

```python
is_active = True
is_admin = False
```

Comparisons in Python, such as numeric or string comparisons, result in boolean values:

```python
>>> 20 > 10
True
>>> 20 < 10
False
>>> 'a' < 'b'
True
>>> 'a' > 'b'
False
```

## The `bool()` Function

To determine if a value is `True` or `False`, you can use the `bool()` function:

```python
>>> bool('Hi')
True
>>> bool('')
False
>>> bool(100)
True
>>> bool(0)
False
```

Some values evaluate to `True`, while others evaluate to `False`.

## Falsy and Truthy Values

Values that evaluate to `True` are truthy, and those that evaluate to `False` are falsy. Falsy values include:

- The number zero (0)
- An empty string `''`
- `False`
- `None`
- An empty list `[]`
- An empty tuple `()`
- An empty dictionary `{}`

Truthy values are any values not classified as falsy.

**Summary:**

- Python boolean data type has two values: `True` and `False`.
- Use the `bool()` function to test if a value is `True` or `False`.
- Falsy values evaluate to `False`, and truthy values evaluate to `True`.
- Falsy values include zero, an empty string, `False`, `None`, an empty list, an empty tuple, and an empty dictionary. Truthy values are those not considered falsy.
